```python
s = (mpimg.imread('mushroom_1100x1100.tif').astype(np.double))[100:,:-100]
dim = s.shape[0]
S = common.fft(s)
print('groundtruth shape: ',s.shape)

a = 0.4
PSF, OTF = preSIM.PSF_OTF(dim, a, 'bessel')

plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(s, cmap='gray')
plt.title('Original')
plt.subplot(1,2,2)
plt.imshow(np.abs(OTF), cmap='gray')
plt.title('OTF')
plt.show()
```

### Create Raw SIM images and Prepare Wiener Filter

The raw SIM images are generated with noise. Copies of the OTF are shifted and absolute and conjugate copies are also created. These copies will be used later for the Wiener filter. 

```python
angle = np.array([0, np.arcsin(4/5), np.pi - np.arcsin(4/5)])
phase = np.array([0, 2*np.pi/3, 4*np.pi/3]) # [0 120 240] 
m = 0.8 # modulating factor
f = 100 # frequency

#Create the noisy raw images
d_h, k = preSIM.raw_SIM_images_w_noise(s, OTF, m, f, angle, phase, 10)
D_H = common.fft(d_h)

OTF_x = np.repeat(np.zeros((2*dim,2*dim),dtype=np.complex128)[np.newaxis], 9, axis=0)
for t in range(9):
    OTF_x[t,dim//2:-dim//2,dim//2:-dim//2] = OTF
OTF_s = postSIM.shift_spectrum(OTF_x, f, angle, 0)
OTFc_s = np.conj(OTF_s)
OTFa_s = np.abs(OTF_s)
```

### SIM Processing

Begin processing SIM images to just right before wiener filtering.

```python
# pad images
D_H_x = postSIM.pad_image_sequence(D_H, 2*dim, 0)
# separate frequency components
D_H_M = postSIM.separate_freq_comp(D_H_x, m, phase)
# Shift frequency components
D_H_M_s = postSIM.shift_spectrum(D_H_M, f, angle, 4)


# Appodization function?
apodize = False

if apodize:
    apod = postSIM.apodization(dim*2, 125, 100, angle)
    D_H_M_s = apod * D_H_M_s
```

### The two Wiener functions used for the reconstruction

$D_p$ are the SIM images

$H_p$ are the shifted OTFs

$\mathcal{J}$ is partioned into $J$ and $J^c$.

$J = \{0_{{\theta_{1}}},0_{\theta_{2}},+k_{\theta_{2}},-k_{\theta_{2}},0_{k_{\theta_{3}}},+k_{\theta_{3}},-k_{\theta_{3}}\}$ and $J^c = \{+k_{\theta_{1}},-k_{\theta_{1}}\}$ where the $0$ is the center component of the angle and $\pm k$ are the shifted components of the angle.

$$f_w(D,H)_J = \frac{\sum_{p=J} H_p^*D_p}{\sum_{p=J} |H_p|^2 + w}$$

$$f_w(D,H)_{J^c} = \frac{\sum_{p=J^c} H_p^*D_p}{\sum_{p=J^c} |H_p|^2 + w}$$

The function $f_w(D,H)_J$ will consist of the majority "petals" with exclusion of one angle. The function $f_w(D,H)_{J^c}$ will be the two "petals" excluded from $f_w(D,H)_J$.


```python
def f_w_j(spectrum, Hc, Ha, w=0, missing_param=None):
    # in:
    # spectrum - spectrum to be filtered
    # Hc - conjugate of the OTF
    # Ha = absolute of the OTF
    # w = wiener parameter
    
    # out: wiener-filtered spectrum
    wiener_num = np.zeros((spectrum.shape[1],spectrum.shape[2]), dtype=np.complex128)
    wiener_denom = np.zeros((spectrum.shape[1],spectrum.shape[2]), dtype=np.complex128)
    
    for i in range(spectrum.shape[0]):
        if (i==missing_param) or (i==missing_param+1):
            continue
        wiener_num += (Hc[i] * spectrum[i])
        wiener_denom += np.power(Ha[i],2)
    wiener_denom += w**2
    return wiener_num / wiener_denom
```

```python
def f_w_j_c(spectrum, Hc, Ha, w=0, missing_param=None):
    # in:
    # spectrum - spectrum to be filtered
    # Hc - conjugate of the OTF
    # Ha = absolute of the OTF
    # w = wiener parameter
    
    # out: wiener-filtered spectrum
    wiener_num = np.zeros((spectrum.shape[1],spectrum.shape[2]), dtype=np.complex128)
    wiener_denom = np.zeros((spectrum.shape[1],spectrum.shape[2]), dtype=np.complex128)
    
    for i in range(spectrum.shape[0]):
        if (i==missing_param) or (i==missing_param+1):
            wiener_num += (Hc[i] * spectrum[i])
            wiener_denom += np.power(Ha[i],2)
    wiener_denom += w**2
    return wiener_num / wiener_denom
```

```python
test_num = 21
gt_w = np.zeros(test_num) # mse of f compared to ground truth
gt_fj = np.zeros(test_num) # mse of g compared to ground truth
fjc_w = np.zeros(test_num) # mse of f compared to noisy image
fjc_fj = np.zeros(test_num) # mse of g compared to noisy image
w = np.linspace(0.01, 0.06, test_num)

for t in range(test_num):
    clear_output(wait=True)
    print('Loop: ' + str(t))
    
    # w(x)
    W_k = postSIM.wiener_filter(D_H_M_s, OTFc_s, OTFa_s, w[t])
    W_k = W_k[dim//2:-dim//2,dim//2:-dim//2]
    w_x = np.abs(np.fft.ifft2(W_k))
    
    # f_w(x)_j
    F_j_k = f_w_j(D_H_M_s, OTFc_s, OTFa_s, w[t], 1)
    F_j_k = F_j_k[dim//2:-dim//2,dim//2:-dim//2]
    f_j_x = np.abs(np.fft.ifft2(F_j_k))
    
    # f_w(x)_j_c (the noisy image)
    F_j_c_k = f_w_j_c(D_H_M_s, OTFc_s, OTFa_s, w[t], 1)
    F_j_c_k = F_j_c_k[dim//2:-dim//2,dim//2:-dim//2]
    f_j_c_x = np.abs(np.fft.ifft2(F_j_c_k))

    # Mean square error - groundtruth
    gt_w[t] = (((np.abs(s)-np.abs(w_x))**2).sum())/(dim**2)
    gt_fj[t] = (((np.abs(s)-np.abs(f_j_x))**2).sum())/(dim**2)
    
    # Mean square error - self-supervised
    fjc_w[t] = (((np.abs(f_j_c_x)-np.abs(w_x))**2).sum())/(dim**2)
    fjc_fj[t] = (((np.abs(f_j_c_x)-np.abs(f_j_x))**2).sum())/(dim**2)
```

```python
fig, ax = plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(w[1:], gt_w[1:], label='w(x) - ground truth')
plt.plot(w[1:], gt_fj[1:]-25, '-', label='f(x) - ground truth')
#plt.axis('tight')
plt.legend()
plt.xlabel('w')
plt.subplot(1,2,2)
plt.plot(w[:-3], fjc_w[:-3], '--', label='w(x) - self-supervised')
plt.plot(w[:-3], fjc_fj[:-3], '--', label='f(x) - self-supervised')
#ax.xaxis.grid(True, which='major')
#ax.xaxis.grid(True, which='minor')
#ax.xaxis.set_minor_locator(MultipleLocator(0.001))
plt.legend()
plt.xlabel('w')
plt.show()

print('Minimums')
print('w(x) - ground truth: ' + str(w[np.argmin(gt_w)]))
print('f(x) - ground truth: ' + str(w[np.argmin(gt_fj)]))
print('w(x) - self-supervised: ' + str(w[np.argmin(fjc_w)]))
print('f(x) - self-supervised: ' + str(w[np.argmin(fjc_fj)]))
```

### The second attempt to using the Noise2Self method while separating the low frequency information. 

$D_p$ are the SIM images

$H_p$ are the shifted OTFs

$\mathcal{J}$ is partioned into $J$ and $J^c$.

$J = \{+k_{\theta_{1}},-k_{\theta_{1}},+k_{\theta_{2}},-k_{\theta_{2}},+k_{\theta_{3}},-k_{\theta_{3}}\}$ and $J^c = \{0_{{\theta_{1}}},0_{{\theta_{2}}},0_{{\theta_{3}}}\}$ where the $0$ is the center component of the angle and $\pm k$ are the shifted components of the angle.

$$f_w(D,H)_J = \frac{\sum_{p=J} H_p^*D_p}{\sum_{p=J} |H_p|^2 + w}$$

$$f_w(D,H)_{J^c} = \frac{\sum_{p=J^c} H_p^*D_p}{\sum_{p=J^c} |H_p|^2 + w}$$

The function $f_w(D,H)_J$ will consist of the outside "petals" with exclusion of the center components. The function $f_w(D,H)_{J^c}$ will be the three center components excluded from $f_w(D,H)_J$.


```python
def wiener_filter_modify_center(spectrum, Hc, Ha, w=0, missing_param=None):
    # in:
    # spectrum - spectrum to be filtered
    # Hc - conjugate of the OTF
    # Ha = absolute of the OTF
    # w = wiener parameter
    
    # out: wiener-filtered spectrum
    wiener_num = np.zeros((spectrum.shape[1],spectrum.shape[2]), dtype=np.complex128)
    wiener_denom = np.zeros((spectrum.shape[1],spectrum.shape[2]), dtype=np.complex128)
    
    for i in range(spectrum.shape[0]):
        if ((i%3)==0):
            continue
        wiener_num += (Hc[i] * spectrum[i])
        wiener_denom += np.power(Ha[i],2)
    wiener_denom += w**2
    return wiener_num / wiener_denom
```

```python
def wiener_filter_leftover_center(spectrum, Hc, Ha, w=0, missing_param=None):
    # in:
    # spectrum - spectrum to be filtered
    # Hc - conjugate of the OTF
    # Ha = absolute of the OTF
    # w = wiener parameter
    
    # out: wiener-filtered spectrum
    wiener_num = np.zeros((spectrum.shape[1],spectrum.shape[2]), dtype=np.complex128)
    wiener_denom = np.zeros((spectrum.shape[1],spectrum.shape[2]), dtype=np.complex128)
    
    for i in range(spectrum.shape[0]):
        if ((i%3)==0):
            wiener_num += (Hc[i] * spectrum[i])
            wiener_denom += np.power(Ha[i],2)
    wiener_denom += w**2
    return wiener_num / wiener_denom
```

```python
gt_w2 = np.zeros(test_num) # mse of f compared to ground truth
gt_fj2 = np.zeros(test_num) # mse of g compared to ground truth
fjc_w2 = np.zeros(test_num) # mse of f compared to noisy image
fjc_fj2 = np.zeros(test_num) # mse of g compared to noisy image
w = np.linspace(0.01, 0.16, test_num)

for t in range(test_num):
    clear_output(wait=True)
    print('Loop: ' + str(t))
    
    # w(x)
    W_k = postSIM.wiener_filter(D_H_M_s, OTFc_s, OTFa_s, w[t])
    W_k = W_k[dim//2:-dim//2,dim//2:-dim//2]
    w_x = np.abs(np.fft.ifft2(W_k))
    
    # f_w(x)_j
    F_j_k = wiener_filter_modify_center(D_H_M_s, OTFc_s, OTFa_s, w[t], 1)
    F_j_k = F_j_k[dim//2:-dim//2,dim//2:-dim//2]
    f_j_x = np.abs(np.fft.ifft2(F_j_k))
    
    # f_w(x)_j_c (the noisy image)
    F_j_c_k = wiener_filter_leftover_center(D_H_M_s, OTFc_s, OTFa_s, w[t], 1)
    F_j_c_k = F_j_c_k[dim//2:-dim//2,dim//2:-dim//2]
    f_j_c_x = np.abs(np.fft.ifft2(F_j_c_k))

    # Mean square error - groundtruth
    gt_w2[t] = (((np.abs(s)-np.abs(w_x))**2).sum())/(dim**2)
    gt_fj2[t] = (((np.abs(s)-np.abs(f_j_x))**2).sum())/(dim**2)
    
    # Mean square error - self-supervised
    fjc_w2[t] = (((np.abs(f_j_c_x)-np.abs(w_x))**2).sum())/(dim**2)
    fjc_fj2[t] = (((np.abs(f_j_c_x)-np.abs(f_j_x))**2).sum())/(dim**2)
```

```python
fig, ax = plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(w[:8], gt_w2[:8], '-', label='w(x) - ground truth')
plt.plot(w[:8], gt_fj2[:8], '-', label='f(x) - ground truth')
#plt.axis('tight')
plt.legend()
plt.xlabel('w')
plt.subplot(1,2,2)
plt.plot(w, fjc_w2, '--', label='w(x) - self-supervised')
plt.plot(w, fjc_fj2, '--', label='f(x) - self-supervised')
#ax.xaxis.grid(True, which='major')
#ax.xaxis.grid(True, which='minor')
#ax.xaxis.set_minor_locator(MultipleLocator(0.001))
plt.legend()
plt.xlabel('w')
plt.show()

print('Minimums')
print('w(x) - ground truth: ' + str(w[np.argmin(gt_w2)]))
print('f(x) - ground truth: ' + str(w[np.argmin(gt_fj2)]))
print('w(x) - self-supervised: ' + str(w[np.argmin(fjc_w2)]))
print('f(x) - self-supervised: ' + str(w[np.argmin(fjc_fj2)]))
```

## Conclusion

Noise2Self was implemented in two methods. The first method was made by patitioning the reconstruction and excluding one angle. The second method was made by excluding the center components from the reconstruction. Comparing the complete wiener filter reconstruction to the ground truth we see the optimal wierner parameter is 0.0345. Comparing $f_w(D,H)_J$ with the ground truth the optimal wiener filter will be 0.0275 for the first method and 0.025 for the second method. For the first method, the self supervised method says that the optimum wiener filter component is at 0.0205 with an error of 0.0045 (18\% error) compared to the optimal wiener filter component. For the second method, we see that the minimum is much further to the left at 0.09, with an error of 2.6 (260\% error).
